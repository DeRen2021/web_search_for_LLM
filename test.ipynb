### Test parse web page with and chunk split

```python
from web_page_parse.parse_web_function import parse_web_page
from similiarity_search.chunck_split import chunk_split
```

```python
web_text_dict = {}
```

```python
import time

web_text_dict = {}
total_parse_time = 0.0  # 解析网页的总时间
total_chunk_time = 0.0  # 文本分块的总时间

for i in r1 + r2:  # 直接合并 r1 和 r2，避免重复代码
    url = i['url']
    
    # 计算 parse_web_page 运行时间
    start_parse = time.time()
    text = parse_web_page(url)
    end_parse = time.time()
    parse_time = end_parse - start_parse
    total_parse_time += parse_time  # 累加解析时间

    # 计算 chunk_split 运行时间
    start_chunk = time.time()
    text_chunks = chunk_split(text)
    end_chunk = time.time()
    chunk_time = end_chunk - start_chunk
    total_chunk_time += chunk_time  # 累加分块时间

    for chunk in text_chunks:
        web_text_dict[chunk] = url

    print(f"URL: {url} - Parsing Time: {parse_time:.4f}s, Chunking Time: {chunk_time:.4f}s")

# 输出总时间
print(f"\nTotal Parsing Time: {total_parse_time:.4f}s")
print(f"Total Chunking Time: {total_chunk_time:.4f}s")
print(f"Total Time (Parsing + Chunking): {total_parse_time + total_chunk_time:.4f}s")
```

```python
for i in web_text_dict:
    print(len(i.split(' ')))
    print(i)
```
